# Importing Modules

```python
from fuzzy_matcher import FuzzyMatcher

import numpy as np
import pandas as pd
```

# Preping a demo example

```python
string_1 = 'Sparsh Dutta'
string_2 = 'Sparsh Dutta rocks!!!'
n1 = 400
n2 = int(n1/8)

base_data = pd.DataFrame()
base_data['sess_id'] = range(n1)
base_data['Combined_Drug'] = [string_1] * n1

comparator_data = pd.DataFrame()
comparator_data['drug_id'] = range(n2)
comparator_data['Asset_Name'] = [string_2] * n2
```

```python
base_data
```

```python
comparator_data
```

# Intiating the Config for Fuzzy Matcher

```python
fuzzy_config = {
                'base' : {
                    'data' : base_data,
                    'identifier' : 'sess_id' ,
                    'column' : 'Combined_Drug'
                },
                'comparator' : {
                    'data' : comparator_data,
                    'identifier' : 'drug_id' ,
                    'column' : 'Asset_Name'
                },
                'top_n' : 2, 
                'metric' : 'jaro-winkler'
                }

# Initiating the class
fm_extractor = FuzzyMatcher(config=fuzzy_config)
# Running the module
fm_extractor.run_accelerator()
```
